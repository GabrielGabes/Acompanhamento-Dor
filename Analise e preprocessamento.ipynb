### Tipos de Escalas de Dor

1. **Numérica Visual**
   - **Descrição:** atendimento avalia a dor numa escala de 0 a 10.
   - **Uso:** Para atendimentos capazes de usar escalas numéricas.

2. **Categórica Numérica**
   - **Descrição:** Dor categorizada em níveis numéricos (leve, moderada, intensa).
   - **Uso:** Quando é preferível categorizar a dor. leve (1-3), moderada (4-6) e intensa (7-10).

3. **PAINAD / ADDA (Demência)**
   - **Descrição:** Avalia dor em atendimentos com demência através de observação comportamental.
   - **Uso:** atendimentos com demência avançada

4. **Behavioral Pain Scale - BPS**
   - **Descrição:** Avalia dor em atendimentos intubados e inconscientes observando expressão facial e movimentos.
   - **Uso:** atendimentos críticos em UTI.

5. **COMFORT-B**
   - **Descrição:** Mede dor/desconforto em crianças e atendimentos incapacitados. Avalia nove categorias de comportamento, incluindo alerta, calma/agitação, e expressão facial.
   - **Uso:** Pediatria e atendimentos incapazes de se comunicar.

6. **Faces Wong-Baker**
   - **Descrição:** atendimento escolhe entre rostos que representam níveis de dor.
   - **Uso:** Pediatria e atendimentos com dificuldade em usar escalas numéricas.

7. **Behavioral Pain Assessment Scale**
   - **Descrição:** Avalia dor com base no comportamento.
   - **Uso:** atendimentos inconscientes ou com comprometimento cognitivo.

8. **FLACC**
   - **Descrição:** Avalia dor através de cinco critérios comportamentais.
   - **Uso:** Crianças pequenas e pacientes incapazes de comunicar dor.

9. **NIPS (Neonatal Infant Pain Scale)**
   - **Descrição:** Avalia dor em neonatos através de sinais comportamentais.
   - **Uso:** Recém-nascidos.

10. **Categórica Verbal**
    - **Descrição:** Dor classificada verbalmente em categorias.
    - **Uso:** Avaliação qualitativa da dor.

```python
# Funções para lidar tempo
def media_timedelta(timedeltas):
    """
    Calcula a média de uma lista de Timedeltas.
    
    Parâmetros:
    timedeltas (list): Lista de objetos pandas.Timedelta.
    
    Retorna:
    pandas.Timedelta: Média dos Timedeltas.
    """
    if len(timedeltas) == 0:
        return pd.Timedelta(0)
    
    total_time = sum(timedeltas, pd.Timedelta(0))
    average_time = total_time / len(timedeltas)
    
    return average_time

def transform_time(delta_tempo):
    # Extraindo horas, minutos e segundos
    hours, remainder = divmod(delta_tempo.seconds, 3600)
    minutes, seconds = divmod(remainder, 60)

    # Formatando como "hh:mm:ss"
    delta_tempo_str = f'{hours:02}:{minutes:02}:{seconds:02}'
    return delta_tempo_str
```
